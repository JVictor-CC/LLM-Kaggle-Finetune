# Load Validation Data

```python
validation_set = pd.read_json('/kaggle/input/mbpp-sanitized-json/sanitized-mbpp.json')
validation_set
```

```python
validation_resp = validation_set['code'].tolist()
```

# Load Data Generated by the Model

## ChatGPT

```python
chatgpt = pd.read_csv('/kaggle/input/generated/final_GPT_3.5_turbo_0125-responses.csv')
```

```python
chatgpt_resp = chatgpt['Response'].tolist()
```

## Deepseek 6.7b finetune

```python
deepseek_coder_6_7b_finetune = pd.read_csv('/kaggle/input/generated/DeepSeek-6.7-base-2.0-responses.csv')
```

```python
deepseek_coder_6_7b_finetune_resp = deepseek_coder_6_7b_finetune['Response'].tolist()
```

## Deepseek 6.7b instruct

```python
deepseek_coder_6_7b_instruct = pd.read_csv('/kaggle/input/generated/DeepSeek6.7b-instruct-responses.csv')
```
